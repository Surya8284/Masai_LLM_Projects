**Task 1 — Explore and Identify Issues**

```python
import pandas as pd

df = pd.read_csv("/content/used_cars_messy.csv")

# Basic exploration
print(df.shape)
print(df.info())
print(df.describe())
```

**Task 2 — Clean the Data**

```python
# 1. Drop rows where target is null
df = df.dropna(subset=['selling_price'])

# 2. Standardize brand column
df['brand'] = df['brand'].str.strip().str.lower()

# 3. Extract numeric values from mileage
df['mileage'] = df['mileage'].str.extract('(\d+\.?\d*)').astype(float)

# 4. Impute missing values (numerical columns)
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

for col in num_cols:
    df[col] = df[col].fillna(df[col].median())

# 5. Remove duplicates
df = df.drop_duplicates()

# Final check
print(df.info())
```

**Task 3 — Compute Baseline MAE**

```python
from sklearn.metrics import mean_absolute_error

# Actual values
y_true = df['selling_price']

# Baseline prediction (mean)
baseline_pred = [y_true.mean()] * len(y_true)

# MAE calculation
mae = mean_absolute_error(y_true, baseline_pred)

print("Baseline MAE:", mae)
```
